SVM Classifier

Equation of the hyperplane:

y = wx -b    where x is the input feature, y is the label, w is the weight and b is the bias of our model

Gradient descent: is an optimization algorthm for minimzing the loss function in various machine learning algorthms .It is used for updating te parameters of the learning model.

(Updated weight)w = w(previous weight) - α*dw      where α is the learning rate   dw(derivative of loss function with respect to the weight)

b = b - α*db

Learning rate:

Is atuning parameter in an optimization algorthm that determines the tep size at each iteration while moving toward a minimum of a loss function.

```python
#importing numpy library
import numpy as np
```

```python
class SVM_Classifier():    #creating an object

 #initiating the hyperparameters
 def __init__ (self, learning_rate,no_of_iterations,lambda_parameter):
   self.learning_rate = learning_rate
   self.no_of_iterations = no_of_iterations
   self.lambda_parameter = lambda_parameter


 #fitting the dataset to our model - svm classifier
 def fit(self,x,y):
   # m is the number of datapoints -->the number of rows
   #n is the number of input features -->the number of columns
   self.m, self.n = x.shape

   #initiating the weight and bias value
   self.w =np.zeros(self.n)


   #initaiating the bias value
   self.b = 0
   self.x = x
   self.y = y

   #implementing the gradient descent algorthm fo optimization
   for i in range(self.no_of_iterations):
     self.update_weights()

    #Finding the new weight values and bias of the gradient descent
 def update_weights(self):
  #label encoding
  y_label = np.where(self.y <= 0, -1,1)

  for index, x_i in enumerate(self.x):

    condition = y_label[index] *np.dot(x_i, self.w) -( self.b) >= 1
    if (condition == True):
      dw = 2*self.lambda_parameter*self.w
      db = 0
    else:
      dw = 2*self.lambda_parameter*self.w - np.dot(x_i,y_label[index])
      db = y_label[index]

    self.w = self.w - self.learning_rate*dw
    self.b = self.b - self.learning_rate*db



#predicting the target column
 def predict(self, x):

  output = np.dot(x,self.w) -self.b
  predicted_labels = np.sign(output)
  y_hat = np.where(predicted_labels <= -1,0,1)

  return y_hat










```
