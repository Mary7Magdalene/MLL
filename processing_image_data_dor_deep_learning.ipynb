Resizing the image using the pillow library

```python
from PIL import Image
```

```python
img  = Image.open('/content/dog.jpg')
img_resize = img.resize((200,200))
```

```python
#saving the resized image
img_resize.save('dog_image_resized.jpg')
```

```python
img_r = mping.imread('/content/dog_image_resized.jpg')
```

```python
img_r_plot = plt.imshow(img_r)
plt.show()
```

Converting RGB images to grayscale

```python
#importing open cv library
import cv2
img = cv2.imread('/content/dog_image_resized.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
type(grayscale_image)
```

```python
grayscale_image.shape
```
